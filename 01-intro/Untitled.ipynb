## Dropping the outliers


```python
# Filter the dataframe to keep only durations between 1 and 60 minutes
filtered_df = df[(df['duration'] >= 1) & (df['duration'] <= 60)]
```

```python
# Calculate the fraction of records left after dropping outliers
fraction_left = len(filtered_df) / len(df)
print(f"Fraction of records left: {fraction_left:.2%}")
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
# Selecting only the pickup and dropoff location IDs and convert them to string
df['PULocationID'] = df['PULocationID'].astype(str)
df['DOLocationID'] = df['DOLocationID'].astype(str)
```

```python
df['PULocationID']
```

```python
# Create a list of dictionaries for each trip with these IDs
data_dicts = df[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
data_dicts
```

```python
# Initialize a DictVectorizer with sparse output
from sklearn.feature_extraction import DictVectorizer
vectorizer = DictVectorizer(sparse=True)

# Fit and transform the data, keeping it in sparse format
feature_matrix = vectorizer.fit_transform(data_dicts)

# Get the dimensionality of the feature matrix
num_columns = feature_matrix.shape[1]
print("Dimensionality of the feature matrix:", num_columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize a Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(feature_matrix, df['duration'])

# Predict on the training data
predictions = model.predict(feature_matrix)

# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(df['duration'], predictions))
print("RMSE on training data:", rmse)
```

```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming the model and vectorizer are already fitted to the January data

# Load the February dataset

df_val = df_feb

# Convert the IDs to strings to ensure they are treated as categorical data
df_val['PULocationID'] = df_val['PULocationID'].astype(str)
df_val['DOLocationID'] = df_val['DOLocationID'].astype(str)

# Compute duration in minutes
df_val['duration'] = (df_val['tpep_dropoff_datetime'] - df_val['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter for realistic durations between 1 and 60 minutes
df_val = df_val[(df_val['duration'] >= 1) & (df_val['duration'] <= 60)]

# Create a list of dictionaries for each trip with these IDs
data_dicts_val = df_val[['PULocationID', 'DOLocationID']].to_dict(orient='records')

# Initialize a DictVectorizer with sparse output
from sklearn.feature_extraction import DictVectorizer
vectorizer = DictVectorizer(sparse=True)

# Fit and transform the data, keeping it in sparse format
feature_matrix = vectorizer.fit_transform(data_dicts_val)


from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize a Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(feature_matrix_val, df_val['duration'])

# Predict on the validation data
predictions_val = model.predict(feature_matrix_val)

# Calculate the RMSE for the validation data
rmse_val = np.sqrt(mean_squared_error(df_val['duration'], predictions_val))
print("RMSE on validation data:", rmse_val)
```
